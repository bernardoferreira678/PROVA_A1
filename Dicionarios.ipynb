Dicionario
Projeto: Cadastro de Alunos
Descrição: Armazena alunos e notas usando um dicionário.
Dicionários

```python
# Dicionário que guarda os alunos
alunos = {}

# Loop principal
while True:
    print("\n--- MENU DE ALUNOS ---")
    print("1 - Cadastrar aluno")
    print("2 - Listar alunos")
    print("3 - Buscar aluno")
    print("4 - Sair")

    opcao = input("Escolha: ")

    # Cadastrar aluno
    if opcao == "1":
        nome = input("Nome do aluno: ")
        nota = float(input("Nota do aluno: "))
        alunos[nome] = nota
        print("Aluno cadastrado!")

    # Listar alunos
    elif opcao == "2":
        print("\nLista de alunos:")
        for nome, nota in alunos.items():
            print(f"{nome}: {nota}")

    # Buscar aluno
    elif opcao == "3":
        nome = input("Digite o nome do aluno: ")
        if nome in alunos:
            print(f"Aluno encontrado: {nome} — Nota: {alunos[nome]}")
        else:
            print("Aluno não encontrado.")

    # Sair
    elif opcao == "4":
        break

    else:
        print("Opção inválida.")
```
