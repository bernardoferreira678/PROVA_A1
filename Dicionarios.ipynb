Dicionario
Projeto: Cadastro de Alunos
Descrição: Armazena alunos e notas usando um dicionário.
Dicionários

```python

alunos = {}

while True:
    print("\n--- MENU DE ALUNOS ---")
    print("1 - Cadastrar aluno")
    print("2 - Listar alunos")
    print("3 - Buscar aluno")
    print("4 - Sair")

    opcao = input("Escolha: ")

    if opcao == "1":
        nome = input("Nome: ")
        nota = float(input("Nota: "))
        alunos[nome] = nota
        print("Aluno cadastrado!")

    elif opcao == "2":
        print("\nAlunos:")
        for nome, nota in alunos.items():
            print(f"{nome}: {nota}")

    elif opcao == "3":
        nome = input("Nome do aluno: ")
        print(alunos.get(nome, "Aluno não encontrado."))

    elif opcao == "4":
        break

    else:
        print("Opção inválida!")
```
